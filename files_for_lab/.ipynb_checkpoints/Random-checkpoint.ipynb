```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

## Categorical Dataset

```python
categorical = pd.read_csv('categorical.csv') 
display(categorical.head())
```

```python
display(categorical.info())
```

```python
# creating instance of one hot encoder
enc = OneHotEncoder()

# assigning numerical values
hot_col = categorical.select_dtypes('object')
for column in hot_col.columns:
    categorical[column] = pd.DataFrame(enc.fit_transform(categorical[[column]]).toarray())
display(categorical.head())
```

## Numerical Dataset

```python
numerical = pd.read_csv('numerical.csv') 
display(numerical.head())
```

```python
display(numerical.info())
```

## Target Dataset

```python
target = pd.read_csv('target.csv') 
display(target.head())
```

```python
display(target.info())
```

```python
display(target['TARGET_B'].nunique())
display(target['TARGET_D'].nunique())
```

```python
y = target['TARGET_B']
target = target.drop(['TARGET_B'], axis=1)
```

## Concatenate categorical and numerical datasets

```python
X = pd.concat([categorical, numerical], axis=1)
display(X.head())
```

```python
# standardize the numerical variables
transformer = StandardScaler().fit(X)
X_standard = transformer.transform(X)
X = pd.DataFrame(X_standard, columns=X.columns)
display(X.head())
```

## Upsampling

```python
# applying SMOTE for upsampling the dataset
smote = SMOTE()
X, y = smote.fit_resample(X, y)
display(y.value_counts())
```

```python
# split data into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(max_depth=2, random_state=0)
cross_val_scores = cross_val_score(clf, X_train, y_train, cv=5)
display(cross_val_scores)
```

```python
display(np.mean(cross_val_scores))
```
